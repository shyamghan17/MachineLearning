### About the Dataset
1- Rank: The ranking of the country or territory based on some criteria (not specified in the provided data).

2- CCA3: The three-letter country code as per the ISO 3166-1 alpha-3 standard.

3- Country/Territory: The name of the country or territory.

4- Capital: The capital city of the country or territory.

5- Continent: The continent to which the country or territory belongs.

6- Population data for the years 2022, 2020, 2015, 2010, 2000, 1990, 1980, and 1970.

7- Area (km²): The land area of the country or territory in square kilometers.

8- Density (per km²): The population density, i.e., the number of people per square kilometer of land area.

9- Growth Rate: The annual population growth rate.

10- World Population Percentage: The percentage of the world's population that the country or territory represents.
