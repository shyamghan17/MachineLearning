```python
from sklearn.cluster import KMeans
kmeans_models = [KMeans(n_clusters=k).fit(credit_data) for k in range (1, 10)]
innertia = [model.inertia_ for model in kmeans_models]
plt.plot(range(1, 10), innertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=23)
kmeans.fit(credit_data)
```

```python
credit_data['cluster_id'] = kmeans.labels_
```

```python
plt.figure(figsize=(10,6))
fig=px.scatter(credit_data,x='ONEOFF_PURCHASES', y='PURCHASES', color='cluster_id', title='Distribution of clusters based on One off purchases and total purchases', trendline="ols")
fig.show()
```

```python
plt.figure(figsize=(10,6))
fig=px.scatter(credit_data,x='CREDIT_LIMIT', y='PURCHASES', color='cluster_id', title='Distribution of clusters based on Credit Limit and total purchases', trendline="ols")
fig.show()
```
