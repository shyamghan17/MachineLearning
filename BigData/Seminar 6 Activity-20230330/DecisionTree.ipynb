```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

irisDataSet = pd.read_csv('IRIS.csv')
irisDataSet
```

```python
features = list(irisDataSet.columns[:4])
x=pd.DataFrame(irisDataSet[['sepal_length', 'sepal_width','petal_length','petal_width']])
y=irisDataSet['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=0)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
plt.title("MACHINE LEARNING ALGORITHMS")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
image = mpimg.imread("ML_Algorithms.jpg")
plt.imshow(image)
plt.show()
```

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
cn=['setosa', 'versicolor', 'virginica']
from sklearn.metrics import classification_report
y_true = y_test
y_pred =classifier.predict(x_test)
print(classification_report(y_true, y_pred, target_names=cn))
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,y_pred)
cf
```

```python
import seaborn as sn
sn.heatmap(cf, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
classifier.predict([[6,3.05,3,1.19]])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy =", accuracy_score(y_test, y_pred))
```
