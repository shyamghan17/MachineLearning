```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x =[1, 3, 4, 5, 6, 7]
y = [6,5,4, 7,8,7]
```

```python
plt.plot(x, y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.xlim(0, 10)
plt.ylim(0,10)
plt.title('this is matplotlib library practice class')
plt.show()
plt.savefig('myfirstplot.jpg')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
x = np.linspace(0,10,11)
y = x *2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) # the main figure
axes1.plot(x,y)

# axes1.set_title('zoomed out')
axes1.set_ylabel('Y-axis')
axes1.set_xlabel('X-axis')
axes1.set_ylim(12,  22)
axes1.set_xlim(12, 22)



axes2 = fig.add_axes([0.3, 0.3, 0.5,0.5,])  #second figure and the four values should in between 0-1

axes2.plot(x,y)
# axes2.set_title('zoomed in')
axes2.set_ylabel('Y-axis')
axes2.set_xlabel('X-axis')
axes2.set_ylim(12,  22)
axes2.set_xlim(12, 22)


plt.show()
```
