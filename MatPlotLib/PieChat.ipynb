```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

slice = [60, 40, 30, 20]
lables = ['JavaScript', 'Python', 'Java', 'Ruby']
colors = ['blue', 'red', 'yellow', 'green']
explode = [0, 0,0.05,0]
plt.pie(slice, labels= lables, colors=colors, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%',  wedgeprops={'edgecolor': '#ffffff'} )
plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```
