```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x =[1, 3, 4, 5, 6, 7]
y = [6,5,4, 7,8,7]
```

```python
plt.plot(x, y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.xlim(0, 10)
plt.ylim(0,10)
plt.title('this is matplotlib library practice class')
plt.show()
plt.savefig('myfirstplot.jpg')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
fig, axes = plt.subplots(nrows= 1, ncols=2)
axes[0].plot(x,y)
axes[1]. plot(y,x)
```

```python
fig, axes= plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
axes[1][1].set_ylabel('you can add anything grabing the axes address')
axes[1][0].set_xlabel('you can add anything grabing the axes address')


# fig.tight_layout()
fig.subplots_adjust(wspace=.5, hspace=0.5)
fig.suptitle('figure label title for hole graph', fontsize = 16)
fig.savefig('figure1.png', dpi = 100)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'red', label = 'x-y', lw = 1.5, ls = '--', marker='*', ms =10, markerfacecolor='red', markeredgecolor = 'blue', markeredgewidth = 1.5)
ax.plot(a,b, color = 'green', label = 'a-b', lw = 5.5, ls = '-.')
ax.legend(loc=(1.2, 0.5))

# lines = ax.plot(x, x+1, color='red', linewidth = 5)
# lines[0].set_dashes([5, 2, 10, 2])
```
