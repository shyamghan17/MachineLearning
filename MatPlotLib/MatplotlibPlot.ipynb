```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')

plt.title('sign and cos representation')
plt.show()
```
