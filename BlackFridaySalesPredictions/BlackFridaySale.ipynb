```python
# bivariate analysis
occupation_plt = df.pivot_table(index='Occupation', values='Purchase', aggfunc= np.mean)
occupation_plt.plot(kind='bar', figsize=(13, 7))
plt.xlabel("Occupation")
plt.ylabel('Purchase')
plt.title("oppaction and purchase analysis")
plt.xticks(rotation=0)
plt.show()
```

```python
Age_Plt = df.pivot_table(index='Age', values='Purchase', aggfunc= np.mean)
Age_Plt.plot(kind='bar', figsize=(13, 7))
plt.xlabel("Age")
plt.ylabel('Purchase')
plt.title("oppaction and purchase analysis")
plt.xticks(rotation=0)
plt.show()
```
