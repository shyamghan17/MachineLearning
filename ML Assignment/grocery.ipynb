```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('grocery-expense-Table 1.csv')
df.tail(10)
```

```python
df['Price'] = df['Price'].str.replace('€','')
df['Clubcard Discount'] = df['Clubcard Discount'].str.replace('€','')
df['Total'] = df['Total'].str.replace('€','')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Total'] = df['Total'].astype(float)
df['Clubcard Discount'] = df['Clubcard Discount'].astype(float)
df['Price'] = df['Price'].astype(float)
```

```python
df.dtypes
```

```python
piechart = df['Source'].value_counts().values
piechart
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# plt.style.use('fivethirtyeight')

lables = ['Tesco', 'Dunnes','Noor','Aldi','EuroGaint','Dealz','Mr Price']
colors = ['blue', 'red', 'yellow', 'green']
explode = [0.05, 0,0,0, 0, 0,0]
plt.pie(piechart, labels= lables, colors=colors, explode=explode, shadow=True, startangle=100, autopct='%1.1f%%',  wedgeprops={'edgecolor': '#000'} )
plt.title("Aadarsh & Sunayana's ecpenditure")
plt.tight_layout()
plt.show()
```
